#Desarrollo Analítico

**$V_a$ :**

$$V_a(R_1 + R_2) - V_1 R_1 - V_2 R_2 = 0$$

**$V_a'$ :**

$$V_a(sC_1 + R_3) - V_1 sC_1 = 0$$

$$V_a = V_1 \frac{sC_1}{sC_1 + R_3}$$

$$V_1 \frac{s2C_1 R_1}{sC_1 + R_3} - V_1 R_1 - V_2 R_1 = 0$$

$$V_1 \frac{s2C_1R_1 - sC_1R_1 - R_1R_3}{sC_1 + R_3} = R_1 V_2$$

*Cancelo $R_1$*

$$V_1 \frac{sC_1 - R_3}{sC_1 + R_3} = V_2$$

$$T(s) = \frac{sC_1 - R_3}{sC_1 + R_3} \qquad ; \qquad T(s) = \frac{s - \frac{R_3}{C_1}}{s + \frac{R_3}{C_1}}$$

$$s\omega_2 = R_3 \qquad ; \qquad R' = 1 \qquad ; \qquad C' = \frac{C_1}{R_3}$$

$$T'(s) = \frac{s - \frac{1}{C'}}{s + \frac{1}{C'}}$$

$$\Omega_\omega = \frac{1}{C'} \qquad ; \qquad \$ = s\Omega_\omega$$

$$T''(\$) = \frac{\$ - 1}{\$ + 1}$$
